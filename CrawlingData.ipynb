```python
def get_page_html(url):
    res = requests.get(url=url, headers=REQUEST_HEADER)
    return res.text
def get_restaurant_name(soup):
    name = soup.find('h1', itemprop='name')
    return name.text.strip() if name else None

def get_restaurant_address(soup):
    address = soup.find('div', class_='res-common-add')
    if address:
        address_full = address.text.strip().replace('\t', '')

        prompt = f"""
            Please separate the following address into 3 parts: street (may have ward), district, city:
            {address_full}

            Ensure that the 3 parts output are Vietnamese
            The "street" only has the house number and proper nouns, delete prefix words or suffix words such as Số, Phố, Đường, Phường, example: 54 Liễu Giai, Cống Vị.
            Ensure that the "district" value is one of the following 12 options: Ba Đình, Cầu Giấy, Đống Đa, Hai Bà Trưng, Hoàn Kiếm, Thanh Xuân, Hoàng Mai, Long Biên, Hà Đông, Tây Hồ, Nam Từ Liêm, Bắc Từ Liêm.
            The "city" value must be Hà Nội.
            For any information not specified in the address return `null`.
            Provide the result in the following JSON format:
            {{
            "street": "...",
            "district": "...",
            "city": "..."
            }}
        """
        response = llm.invoke(prompt)
        response_text = str(response.content)
        try:
            result_dict = json.loads(re.search(r'\{.*?\}', response_text, re.DOTALL).group(0))
            return result_dict
        except (json.JSONDecodeError, AttributeError):
            return {'street': None, 'district': None, 'city': None}


from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time

def get_restaurant_location(address):
    addr = f"{address['street']}, {address['district']}, {address['city']}"
    i = 0
    latitude = None
    longitude = None
    count = 0
    while latitude is None or longitude is None:
        try:
            geolocator = Nominatim(user_agent=f"test{i}")
            location = geolocator.geocode(addr)
            if location:
                latitude = location.latitude
                longitude = location.longitude
        except Exception as e:
            count += 1
            print(f"count: {count}")
            if count >= 10:
                i += 1
                print(f"i: {i}")
                count = 0
        finally:
            if latitude is not None and longitude is not None:
                break
    return {'latitude': latitude, 'longitude': longitude}

def get_restaurant_rating(soup):
    rating = soup.find('div', itemprop='ratingValue', class_='microsite-point-avg')
    return rating.get_text(strip=True) if rating else None

def get_restaurant_time(soup):
    # Tìm div chứa thông tin về thời gian
    time_div = soup.find('div', class_='micro-timesopen')

    if time_div:
        # Tìm span chứa thời gian thực tế
        time_span = time_div.find_all('span')[2]  # Index 2 nhắm vào span chứa thời gian

        if time_span:
            time_text = time_span.get_text(strip=True)
            # Tách thời gian mở và đóng cửa (vd: '10:00 - 23:00')
            time_parts = time_text.split(' - ')
            if len(time_parts) == 2:
                start_time, end_time = time_parts
                # Trả về một dictionary với định dạng TIME
                return {
                    "start_time": start_time.strip(),
                    "end_time": end_time.strip()
                }

    # Trả về giá trị mặc định nếu không có thông tin thời gian
    return {
        "start_time": None,
        "end_time": None
    }


def get_restaurant_description(soup):
    # Extract cuisine type
    cuisine = soup.find('div', itemprop='servesCuisine')
    cuisine_text = cuisine.get_text(strip=True) if cuisine else ""

    # Extract audience
    audience = soup.find('div', class_='audiences')
    audience_text = audience.get_text(strip=True).replace('&nbsp;', ' ') if audience else ""

    # Extract category
    category = soup.find('div', class_='category-items')
    category_text = category.get_text(strip=True) if category else ""

    # Combine all parts into a single description
    description = "Đây là " + category_text + " phù hợp với " + audience_text + ", kiểu " + cuisine_text + "."

    return  description

def get_restaurant_comments(soup):
    comments = []
    comment_elements = soup.findAll('div', class_='rd-des')
    for comment in comment_elements:
        text = comment.text.strip()

        # Loại bỏ phần "Xem thêm" và ký tự xuống dòng '\n'
        clean_text = text.replace('...Xem thêm', '').replace('\n', '')

        comments.append(clean_text)
    return comments


import json

def get_restaurant_price_range(soup):
    # Giả sử khoảng giá nằm trong một thẻ div với class 'res-common-minmaxprice'
    price_element = soup.find('div', class_='res-common-minmaxprice')
    if not price_element:
        return None  # Trả về None nếu không tìm thấy phần tử

    price_range = price_element.text.strip()

    # Loại bỏ ký hiệu không phải số và dấu phân cách hàng nghìn
    # Chỉ giữ lại các ký tự số và dấu gạch ngang
    clean_price_range = ''.join(c for c in price_range if c.isdigit() or c == '-')

    # Khởi tạo giá trị mặc định
    min_price = None
    max_price = None
    fixed_price = None

    # Tách giá trị min và max
    if '-' in clean_price_range:
        parts = clean_price_range.split('-')
        if len(parts) == 2:
            try:
                min_price = int(parts[0].strip())
                max_price = int(parts[1].strip())
            except ValueError:
                print(f"Cảnh báo: Định dạng giá không hợp lệ: {clean_price_range}")
    else:
        # Nếu không có dấu `-`, coi giá trị là giá cố định
        try:
            fixed_price = int(clean_price_range.strip())
        except ValueError:
            print(f"Cảnh báo: Giá không thể chuyển đổi thành số: {clean_price_range}")

    # Tạo đối tượng dictionary
    price_info = {}
    if min_price is not None and max_price is not None:
        price_info['min_price'] = min_price
        price_info['max_price'] = max_price
    elif fixed_price is not None:
        price_info['fixed_price'] = fixed_price

    # Chuyển đổi dictionary thành chuỗi JSON
    json_string = json.dumps(price_info, ensure_ascii=False)

    # Trả về chuỗi JSON
    return json_string

def get_restaurant_type(soup, types):
    """Trích xuất loại hình nhà hàng từ trang Foody.vn và trả về dạng chuỗi văn bản"""

    # Duyệt qua tất cả các phần tử <a> trong trang
    a_elements = soup.find_all('a')
    found_types = set()  # Sử dụng set để lưu loại hình không trùng lặp

    for element in a_elements:
        text = element.get_text(strip=True)
        if text in types:
            found_types.add(text)

    # Chuyển set thành chuỗi văn bản với dấu phân cách là dấu phẩy
    result_text = ', '.join(found_types)
    return result_text

# Example list of restaurant types
restaurant_types = ["Nhà hàng", "Quán ăn", "Café/Dessert", "Tiệm bánh", "Bar", "Sang trọng", "Buffet", "Karaoke", "Tiệc cưới/Hội nghị", "Ăn chay"]

def get_restaurant_img(soup):
    div = soup.find('div', class_='img')
    img_tag = div.findAll('img')
    img_url = [img['src'] for img in img_tag]

    # Create a dictionary
    img_dict = {'img_url': img_url}

    # Convert the dictionary to a JSON string
    json_string = json.dumps(img_dict, ensure_ascii=False)

    return json_string

def get_restaurant_suitable_for(soup, types):
    """Trích xuất tất cả các loại hình nhà hàng từ trang Foody.vn"""

    # Duyệt qua tất cả các phần tử <a> trong trang
    a_elements = soup.find_all('a')
    found_types = []  # Danh sách chứa tất cả các loại hình phù hợp

    for element in a_elements:
        text = element.get_text(strip=True)
        if text in types:
            found_types.append(text)

    # Chuyển danh sách thành chuỗi văn bản với dấu phân cách là dấu phẩy
    result_text = ', '.join(found_types)
    return result_text


# Danh sách các loại hình nhà hàng
suitable_for_types = ["Hẹn hò", "Họp nhóm", "Thư giãn", "Đãi tiệc", "Ăn gia đình", "Tiếp khách", "Ngắm cảnh" , "Nghe nhạc", "Buffet", "Ăn chay", "Tiệc ngoài trời", "Uống bia - Nhậu", "Du lịch", "BBQ - Món Nướng", "Ăn Fastfood", "Ăn vặt", "Nghe nhạc", "Chụp hình - Quay phim"]

def has_kids_play_area(soup):
    """Kiểm tra xem có chỗ chơi cho trẻ em, dựa trên màu chữ"""

    # Tìm tất cả các thẻ <a> hoặc <span> chứa thông tin "Có chỗ chơi cho trẻ em"
    span_element = soup.find('span', string="Có chỗ chơi cho trẻ em")
    a_element = soup.find('a', string="Có chỗ chơi cho trẻ em")

    # Kiểm tra màu sắc và trả về giá trị boolean
    if a_element and 'font-weight:bold' in a_element.get('style', ''):
        return True  # Màu xanh

    if span_element and 'color:#aaa' in span_element.get('style', ''):
        return False  # Màu xám

    return None  # Nếu không tìm thấy phần tử phù hợp


def has_parking_available(soup):
    """Kiểm tra xem có chỗ chơi cho trẻ em, dựa trên màu chữ"""

    # Tìm tất cả các thẻ <a> hoặc <span> chứa thông tin "Có chỗ chơi cho trẻ em"
    span_element = soup.find('span', string="Có chỗ đậu ôtô")
    a_element = soup.find('a', string="Có chỗ đậu ôtô")

    # Kiểm tra màu sắc và trả về giá trị boolean
    if a_element and 'font-weight:bold' in a_element.get('style', ''):
        return True  # Màu xanh

    if span_element and 'color:#aaa' in span_element.get('style', ''):
        return False  # Màu xám

    return None  # Nếu không tìm thấy phần tử phù hợp


def extract_restaurant_info(url):
    info = {}
    html = get_page_html(url=url)
    soup = bp(html, 'lxml')
    info['name'] = get_restaurant_name(soup)
    info['address'] = get_restaurant_address(soup)
    info['location'] = get_restaurant_location(info['address'])
    info['rating'] = get_restaurant_rating(soup)
    info['working_hour'] = get_restaurant_time(soup)
    info['restaurant_type'] = get_restaurant_type(soup, restaurant_types)
    info['description'] = get_restaurant_description(soup)
    info['embedding_description '] = get_embedding(info['description'])
    info['suitable_for'] = get_restaurant_suitable_for(soup, suitable_for)
    info['kids_play_area'] = has_kids_play_area(soup)
    info['parking_available'] = has_parking_available(soup)
    info['price_range'] = get_restaurant_price_range(soup)
    info['img_url'] = get_restaurant_img(soup)
    info['comments'] = get_restaurant_comments(soup)
    return info
```

```python
import psycopg2
import json

def insert_restaurant(restaurant_info):
    conn = psycopg2.connect(postgresql_url)
    cur = conn.cursor()

    # Debugging: Print the restaurant_info to inspect its contents
    print("Restaurant Info:", restaurant_info)

    # Extracting values from restaurant_info with checks for None values
    name = restaurant_info.get('name')
    street = restaurant_info.get('address', {}).get('street')
    district = restaurant_info.get('address', {}).get('district')
    city = restaurant_info.get('address', {}).get('city')
    latitude = restaurant_info.get('location', {}).get('latitude')
    longitude = restaurant_info.get('location', {}).get('longitude')
    start_time = restaurant_info.get('working_hour', {}).get('start_time')
    end_time = restaurant_info.get('working_hour', {}).get('end_time')
    suitable_for = restaurant_info.get('suitable_for')
    if suitable_for:  # Convert list to array for PostgreSQL
        suitable_for = '{' + ','.join(suitable_for) + '}'
    else:
        suitable_for = None
    restaurant_type = restaurant_info.get('restaurant_type')
    if restaurant_type:  # Convert list to array for PostgreSQL
        restaurant_type = '{' + ','.join(restaurant_type) + '}'
    else:
        restaurant_type = None
    rating = restaurant_info.get('rating')
    description = restaurant_info.get('description')
    embedding_description = restaurant_info.get('embedding_description')
    if embedding_description:  # Convert list to array for PostgreSQL vector
        embedding_description = '{' + ','.join(map(str, embedding_description)) + '}'
    else:
        embedding_description = None
    parking_available = restaurant_info.get('parking_available')
    kids_play_area = restaurant_info.get('kids_play_area')
    img_url = json.dumps(restaurant_info.get('img_url'))
    comments = json.dumps(restaurant_info.get('comments'))
    price_range = json.dumps(restaurant_info.get('price_range'))
    # Debugging: Print the values to be inserted
    print("Values to be inserted:", name, street, district, city, latitude, longitude, start_time, end_time, price_range, suitable_for, restaurant_type, rating, description, embedding_description, parking_available, kids_play_area, img_url, comments)

    sql = """
        INSERT INTO Restaurant (
            name, address, location, working_hour, suitable_for, restaurant_type, rating, description, embedding_description, price_range, parking_available, kids_play_area, img_url, comments
        )
        VALUES (
          %s,
          ROW(%s, %s, %s),  -- Address (street, city, zip)
          ST_GeomFromText(%s, 4326),  -- Location (POINT)
          ROW(%s, %s),  -- Working hours (start_time, end_time)
          %s,  -- suitable_for
          %s,  -- restaurant_type
          %s,  -- rating
          %s,  -- description
          %s,  -- price_range
          %s,  -- parking_available
          %s,  -- kids_play_area
          %s,  -- img_url
          %s   -- comments
        )
        """

    # Chỉnh lại việc truyền tham số khớp với số lượng %s
    try:
        cur.execute(sql, (
            name,
            street, district, city,
            longitude, latitude,  # Note the order: longitude, latitude
            start_time, end_time,
            suitable_for, restaurant_type,
            rating, description, embedding_description,
            price_range,  # Truyền `price_range` vào đúng vị trí
            parking_available, kids_play_area,
            img_url, comments
        ))
        conn.commit()
    except Exception as e:
        print("Error inserting data:", e)
        conn.rollback()

```
